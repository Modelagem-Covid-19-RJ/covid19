```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly
import json
import datetime as dt
import matplotlib.dates as mdates
import unidecode
from IPython.display import Image
from PIL import Image, ImageDraw
#import imageio
from get_data import read_data, get_data
import plotly.graph_objects as go

df = pd.read_csv('Dados_MRJ_03-05.csv')
dt_start = '15-03-2020'
dt_fim = '03-05-2020'
bairros = df['Bairro'].unique()
dic_acumulados = {}
dic_novos = {}

for b in bairros:
    if b != 'FORA DO MUNICÍPIO' and b != 'INDEFINIDO':
        dic_acumulados.update({b:get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[0]})
        dic_novos.update({b:get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[1]})
            
data_start = dt_start
data_end = dt_fim
start = dt.datetime.strptime(data_start, "%d-%m-%Y")
then = dt.datetime.strptime(data_end, "%d-%m-%Y")
days = mdates.drange(start,then,dt.timedelta(days=1))
dias = [str(mdates.num2date(v)).split(' ')[0] for v in days]

dic_acumulados.update({'Datas':dias})
dic_novos.update({'Datas':dias})

df_series_temp_acumulados = pd.DataFrame(dic_acumulados)
df_series_temp_novos = pd.DataFrame(dic_novos)

df_series_temp_acumulados.to_csv('dados-series-temp-acumul-bairros.csv')
df_series_temp_novos.to_csv('dados-series-temp-novos-bairros.csv')
```
