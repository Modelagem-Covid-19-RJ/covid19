```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly
import json
import datetime as dt
import matplotlib.dates as mdates
import unidecode
from IPython.display import Image
from PIL import Image, ImageDraw
#import imageio
from get_data import read_data, get_data
import plotly.graph_objects as go

df = read_data(pd.read_csv('Dados_indiv_MRJ_covid19 (2).csv', encoding = 'iso-8859-1', delimiter = ';'), fonte='prefeitura')
dt_start = '15-03-2020'
dt_fim = '15-05-2020'
bairros = df['Bairro'].unique()
dic_acumulados = {}
dic_novos = {}

for b in bairros:
    if b != 'FORA DO MUNICÍPIO' and b != 'INDEFINIDO':
        dic_acumulados.update({b:get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[0]})
        dic_novos.update({b:get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[1]})
            
data_start = dt_start
data_end = dt_fim
start = dt.datetime.strptime(data_start, "%d-%m-%Y")
then = dt.datetime.strptime(data_end, "%d-%m-%Y")
days = mdates.drange(start,then,dt.timedelta(days=1))
dias = [str(mdates.num2date(v)).split(' ')[0] for v in days]

dic_acumulados.update({'Datas':dias})
dic_novos.update({'Datas':dias})

df_series_temp_acumulados = pd.DataFrame(dic_acumulados)
df_series_temp_novos = pd.DataFrame(dic_novos)

df_series_temp_acumulados.to_csv('dados-series-temp-acumul-bairros.csv')
df_series_temp_novos.to_csv('dados-series-temp-novos-bairros.csv')
```

```python
df = read_data(pd.read_csv('Dados_indiv_MRJ_covid19 (2).csv', encoding = 'iso-8859-1', delimiter = ';'), fonte='prefeitura')
dt_start = '15-03-2020'
dt_fim = '15-05-2020'
dic_acumulados_aps = {}
dic_novos_aps = {}

for i, ap in enumerate(aps):
    if ap != 'N/D':
        dic_acumulados_aps.update({f'AP %s' %(i+1):np.sum(np.array([get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[0] for b in ap]), axis = 0)})
        dic_novos_aps.update({f'AP %s' %(i+1):np.sum(np.array([get_data(local = b, df = df,fonte = 'prefeitura', 
                                T_start = dt_start, T_fim = dt_fim, to_print=False)[1] for b in ap]), axis = 0)})
            
start = dt.datetime.strptime(dt_start, "%d-%m-%Y")
then = dt.datetime.strptime(dt_fim, "%d-%m-%Y")
days = mdates.drange(start,then,dt.timedelta(days=1))
dias = [str(mdates.num2date(v)).split(' ')[0] for v in days]

dic_acumulados_aps.update({'Datas':dias})
dic_novos_aps.update({'Datas':dias})

df_series_temp_acumulados_aps = pd.DataFrame(dic_acumulados_aps)
df_series_temp_novos_aps = pd.DataFrame(dic_novos_aps)

df_series_temp_acumulados_aps.to_csv('dados-series-temp-acumul-aps.csv')
df_series_temp_novos_aps.to_csv('dados-series-temp-novos-aps.csv')
```

## Bairros

```python
media_movel_novos = []
media_movel_acumulados = []
for col in df_series_temp_novos.columns[:-1]:
    temp_df_novos = df_series_temp_novos[col].values
    temp_df_acumulados = df_series_temp_acumulados[col].values
    media_movel_novos_b = []
    media_movel_acumulados_b = []
    for i in range(len(temp_df_novos)-2):
        if i < 5:
            media_movel_novos_b += [np.sum(temp_df_novos[:i+2])/(6*temp_df_acumulados[-1])]
            media_movel_acumulados_b += [np.sum(temp_df_acumulados[:i+2])/(6*temp_df_acumulados[-1])]
        else:
            media_movel_novos_b += [np.sum(temp_df_novos[i-4:i+2])/(6*temp_df_acumulados[-1])]
            media_movel_acumulados_b += [np.sum(temp_df_acumulados[i-4:i+2])/(6*temp_df_acumulados[-1])]
    media_movel_novos += [media_movel_novos_b]
    media_movel_acumulados += [media_movel_acumulados_b]    
```

## APs

```python
media_movel_novos_aps = []
media_movel_acumulados_aps = []
for col in df_series_temp_novos_aps.columns[:-1]:
    temp_df_novos_aps = df_series_temp_novos_aps[col].values
    temp_df_acumulados_aps = df_series_temp_acumulados_aps[col].values
    media_movel_novos_aps_ap = []
    media_movel_acumulados_aps_ap = []
    for i in range(len(temp_df_novos_aps)-2):
        if i < 5:
            media_movel_novos_aps_ap += [np.sum(temp_df_novos_aps[:i+2])/(6*temp_df_acumulados_aps[-1])]
            media_movel_acumulados_aps_ap += [np.sum(temp_df_acumulados_aps[:i+2])/(6*temp_df_acumulados_aps[-1])]
        else:
            media_movel_novos_aps_ap += [np.sum(temp_df_novos_aps[i-4:i+2])/(6*temp_df_acumulados_aps[-1])]
            media_movel_acumulados_aps_ap += [np.sum(temp_df_acumulados_aps[i-4:i+2])/(6*temp_df_acumulados_aps[-1])]
    media_movel_novos_aps += [media_movel_novos_aps_ap]
    media_movel_acumulados_aps += [media_movel_acumulados_aps_ap]    
```
